### 1.	Title 
### 2.	Ratings 
### 3.	Price 
 


```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/")
```

```python
search_product = driver.find_element(By.XPATH, "//div[@class='nav-search-field ']//input")
search_product.send_keys('Laptop')


search_btn = driver.find_element(By.CLASS_NAME,"nav-right")
search_btn.click()
```

```python
i7 = driver.find_element(By.XPATH,"//li[@aria-label='Intel Core i7']//div[@class='a-checkbox a-checkbox-fancy s-navigation-checkbox aok-float-left']")

i7.click()
```

```python
title =[]
ratings = []
price=[]


for i in driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal"]'):
    title.append(i.text)
for i in driver.find_elements(By.CLASS_NAME, "a-price-whole"):
    price.append(i.text)
for i in driver.find_elements(By.XPATH, '//div[contains(@class, "s-result-item s-asin")]'):
    ratings_box = i.find_elements(By.XPATH, './/div[@class="a-row a-size-small"]/span')

    if ratings_box != []:
        rating = ratings_box[0].get_attribute('aria-label')
    else:
        rating = 'null'
    
    ratings.append(rating)
```

```python
laptop = pd.DataFrame({'Title':title[:10], 'Price':price[:10], 'Ratings':ratings[:10]}) 
laptop
```

#### The above task will be done in following steps: 
### 1. First get the webpagehttps://www.azquotes.com/ 2. Click on Top Quotes 
### 3. Than scrap a) Quote b) Author c) Type Of Quotes 


```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

```

```python
driver = webdriver.Chrome
```

```python
driver.get('https://www.azquotes.com/')
```

```python
menu = driver.find_elements(By.XPATH,'//div[@class="mainmenu"]/ul/li')
topqt_btn = menu[4]
topqt_btn.click()

title=[]

author =[]

tags= []

for i in driver.find_elements(By.XPATH, "//ul[@class='list-quotes']/li/div/p/a[@class='title']"):
    title.append(i.text)

for i in driver.find_elements(By.XPATH, "//ul[@class='list-quotes']/li/div/div/a"):
    author.append(i.text)

for i in driver.find_elements(By.CLASS_NAME, 'tags'):
    tags.append(i.text)

TQ = pd.DataFrame({"Title":title, 'Author':author, 'Type_of_Quote':tags})
TQ
```
