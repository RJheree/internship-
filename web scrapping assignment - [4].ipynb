```python
 search_2 =driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
search_2.click()
```

```python
 search_3 =driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
search_3.click()
```

```python
url1 =[]
for i in driver.find_elements(By.XPATH,'//a[@class="Link"]'):
    url1.append(i.get_attribute("href"))
```

```python
language=[] 
try:
    for i in driver.find_elements(By.XPATH,'//span[@itemprop="programmingLanguage"]'):
        language.append(i.text)
except NoSuchElementException: 
        language.append('No details available')
```

```python
contributors = []
title = []
description=[]

for url in url1:
    driver.get(url)
    
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/div/main/div/div[1]/div[1]/div/span[1]/a'):
            title.append(i.text)
    except NoSuchElementException: 
        title.append('No details available')
    try:
        for i in driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[1]/div/p'):
            description .append(i.text)
    except NoSuchElementException: 
        description.append('No details available')

        xpath_patterns = [
        '//div[@class="Layout-sidebar"]/div/div[5]/div/h2/a/span',
        '//div[@class="Layout-sidebar"]/div/div[6]/div/h2/a/span',
        '//div[@class="Layout-sidebar"]/div/div[4]/div/h2/a/span',
        '//div[@class="Layout-sidebar"]/div/div[3]/div/h2/a/span'
    ]

    found = False  

    for xpath_pattern in xpath_patterns:
        try:
            elements = driver.find_elements(By.XPATH, xpath_pattern)
            if elements:
                found = True
                contributors.append(elements[0].text)  
                break  
        except NoSuchElementException as e:
            contributors.append('-')
            pass

    if not found:
        contributors.append('-')
```

```python
trending_GitHub = pd.DataFrame({'Repository title':title,'Repository description':description,
                                'Language used':language,'Contributors Count':contributors})
trending_GitHub 

```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu')
driver.maximize_window() 
```

```python
driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
 search_1 =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/label/select/option[5]')
search_1.click()
```

```python
name = []
dtype = []
task = []
attribute = []
no_instances = []
no_attribute = []
year = []

num_next_buttons = 26

for _ in range(num_next_buttons):
    driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]').click()

    try:
        for i in driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]'):
            name.append(i.text)
    except NoSuchElementException: 
        name.append('No details available')

    try:
        for i in driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span'):
            dtype.append(i.text)
    except NoSuchElementException: 
        dtype.append('No details available')

    try:
        for i in driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span'):
            task.append(i.text)
    except NoSuchElementException: 
        task.append('No details available')

    try:
        for i in driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span'):
            no_attribute.append(i.text)
    except NoSuchElementException: 
        no_attribute.append('No details available')
    
    try:
        for i in driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span'):
            no_instances.append(i.text)
    except NoSuchElementException: 
        no_instances.append('No details available')


    try:
        for i in driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]'):
            attribute.append(i.text)
    except NoSuchElementException:
        attribute.append('No details available')
    
    time.sleep(2)

    try:
        for i in driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]'):
            year.append(i.text)
    except NoSuchElementException: 
        year.append('No details available')

    driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]').click()
    
    if _ == 25:  
        break

    try:
        next_button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
        if 'disabled' in next_button.get_attribute('class'):
            break
        time.sleep(2)
        next_button.click()
        time.sleep(5)
        
        
    except NoSuchElementException as e:
        break
```

```python
dataset_UCI=pd.DataFrame({'Dataset name':name,'Data type':dtype,'Task':task,'Attribute type':attribute,'No of instances':no_instances,
                           ' No of attribute':no_attribute,'Year':year})
dataset_UCI

```
